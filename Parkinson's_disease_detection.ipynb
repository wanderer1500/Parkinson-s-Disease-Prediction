```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
parkinsons_data = pd.read_csv("/content/parkinsons.csv")
```

```python
#getting some statistical measures about the data
parkinsons_data.describe()
```

```python
#distribution of target variable (status)
parkinsons_data['status'].value_counts()
```

```python
#grouping the data based on the target column or variable
parkinsons_data.groupby('status').mean(numeric_only=True)
```

```python
X = parkinsons_data.drop(columns = ['status','name'],axis = 1)
Y = parkinsons_data['status']
# print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state = 2)
# print(X_test)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
# print(X_train)
X_test = scaler.transform(X_test)
```

```python
model = svm.SVC(kernel ='linear') #support vector classifier other is regressor
```

```python
model.fit(X_train,Y_train)
```

```python
#model Evaluation
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)
print(training_data_accuracy)
```

```python
X_test_prediction  = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test,X_test_prediction)
print(testing_data_accuracy)
```
